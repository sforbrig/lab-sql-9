```python
# Create a Python connection with SQL database and retrieve the results of the following queries as dataframes
connection_string = 'mysql+pymysql://root:'+password+'@localhost/sakila'
engine = create_engine(connection_string)
```

```python
# How many distinct (different) actors' last names are there?
data = pd.read_sql_query('SELECT COUNT(DISTINCT(last_name)) AS different_actor_names FROM actor', engine)
data.head()
```

```python
# 2. Add an additional column day_type with values 'weekend' and 'workday' depending on the rental day of the week.
# use % twice!!
query = '''SELECT *
,DATE_FORMAT(rental_date, '%%M') AS 'month'
,DATE_FORMAT(rental_date, '%%W') AS 'weekday'
,CASE
WHEN DATE_FORMAT(rental_date, '%%W') > 5 THEN 'weekend'
ELSE 'workday' 
END AS day_type
FROM rental
ORDER BY rental_date;'''

data = pd.read_sql_query(query, engine)
data.head()
```

```python
# 3. Get all films with ARMAGEDDON in the title.
query= 'SELECT * FROM film WHERE title LIKE "%%ARMAGEDDON%%"'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
# 4. Get 10 the longest films.
query= 'SELECT * FROM film ORDER BY length DESC LIMIT 10'
data = pd.read_sql_query(query, engine)
data.head()
```

```python
# 5. How many films include Behind the Scenes content?
query= 'SELECT * FROM film WHERE special_features LIKE "%%Behind the Scenes%%"'
bh_sc = pd.read_sql_query(query, engine)
bh_sc.head()
```

```python
# 6. Which kind of movies (rating) have a mean duration of more than two hours?
query= 'SELECT rating, ROUND(AVG(length),2) as "AVG length" FROM film GROUP BY rating HAVING AVG(length)>120'
hrs_2 = pd.read_sql_query(query, engine)
hrs_2.head()
```

```python
# 7. Rank films by length (filter out the rows that have nulls or 0s in length column). In your output, only select the columns title, length, and the rank.
query= 'SELECT title, length,RANK () OVER (ORDER BY length DESC) AS "rank" FROM film WHERE length IS NOT NULL'
length = pd.read_sql_query(query, engine)
length.head()
```
